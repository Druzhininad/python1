```python
# libs
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction import DictVectorizer
```

```python
# read data
boys = pd.read_csv("Boys.csv", encoding='Windows-1251', sep=';', header=0)
boys.drop('ID', axis=1, inplace=True)
girls = pd.read_csv("Girls.csv", encoding='Windows-1251', sep=';')
girls.drop('ID', axis=1, inplace=True)
```

```python
boys.head()
girls.head()
```

```python
girls['Sex'] = 0
boys['Sex'] = 1
```

```python
df_names = pd.concat([girls, boys])
```

```python
df_names
```

```python
Xfeatures = df_names['Name']
```

```python
cv = CountVectorizer()
X = cv.fit_transform(Xfeatures)
```

```python
cv.get_feature_names()
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = df_names['Sex']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(X_train,y_train)
clf.score(X_test,y_test)
```

```python
print("Accuracy:",clf.score(X_test,y_test))
```

```python
names = ["Дарья","Миша","Степа","Соня","Вера","Рома","Никита"]
```

```python
def appeal(name):
    if clf.predict(cv.transform([name]).toarray()) == 0:
        print("Уважаемая", name)
    else:
        print("Уважаемый", name)
```

```python
for name in names:
    appeal(name)
```
